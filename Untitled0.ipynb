```python
import math

repe = 1

while repe != 2:
    print("----------------------------------------------------------------------")
    print("                                 MENÚ")
    print("----------------------------------------------------------------------")
    print(" OPCIONES")
    print("(1) SUMA")
    print("(2) RESTA")
    print("(3) DIVISIÓN")
    print("(4) MULTIPLICACIÓN")
    print("(5) RAÍZ CUADRADA")
    print("(6) ELEVACIÓN")
    print("(7) SENO Y ARCO")
    print("(8) COSENO Y ARCO")
    print("(9) TANGENTE Y ARCO")
    print("(10) ECUACIÓN DE SEGUNDO GRADO")
    print("----------------------------------------------------------------------")
    print("Digite una opción numérica")
    opcion = int(input())

    if opcion == 1:
        # Operación de suma
        resSuma = 0
        print("Ingrese la cantidad de números a digitar")
        contSuma = int(input())
        for i in range(1, contSuma + 1):
            print("Ingrese un dígito")
            numSuma = int(input())
            resSuma += numSuma
        print("La suma es igual a:", resSuma)

    elif opcion == 2:
        # Operación de resta
        resResta = 0
        print("Ingrese la cantidad de números a digitar")
        contResta = int(input())
        for i in range(1, contResta + 1):
            print("Ingrese un dígito")
            numResta = int(input())
            resResta -= numResta
        print("La resta es igual a:", resResta)

    elif opcion == 3:
        # Operación de división
        print("Ingrese primer dígito")
        numDiv1 = int(input())
        print("Ingrese segundo dígito")
        numDiv2 = int(input())
        resDiv = numDiv1 / numDiv2
        print("La división es igual a:", resDiv)

    elif opcion == 4:
        # Operación de multiplicación
        resMult = 1
        print("Ingrese la cantidad de números a digitar")
        contMult = int(input())
        for i in range(1, contMult + 1):
            print("Ingrese un dígito")
            numMult = int(input())
            resMult *= numMult
        print("La multiplicación es igual a:", resMult)

    elif opcion == 5:
        # Operación de raíz cuadrada
        print("Ingrese radicando")
        numRaiz = float(input())
        resRaiz = math.sqrt(numRaiz)
        print("La raíz cuadrada es igual a:", resRaiz)

    elif opcion == 6:
        # Operación de elevación
        print("Ingrese número")
        numElevacion = float(input())
        print("Ingrese potencia")
        contElevacion = int(input())
        resElevacion = math.pow(numElevacion, contElevacion)
        print("La elevación es igual a:", resElevacion)

    elif opcion == 7:
        # Operación de seno y arco
        print("Seleccione qué operación desea")
        print("(1) Seno")
        print("(2) Arco Seno")
        opSeno = int(input())

        if opSeno == 1:
            # Seno
            print("Ingrese número")
            numSeno = float(input())
            resSeno = math.sin(numSeno)
            print("El seno es igual a:", resSeno)

        elif opSeno == 2:
            # Arco Seno
            print("Ingrese número")
            numArcSeno = float(input())
            resArcSeno = math.asin(numArcSeno)
            print("El arco seno es igual a:", resArcSeno)

        else:
            print("Opción no válida")

    elif opcion == 8:
        # Operación de coseno y arco
        print("Seleccione qué operación desea")
        print("(1) Coseno")
        print("(2) Arco Coseno")
        opCoseno = int(input())

        if opCoseno == 1:
            # Coseno
            print("Ingrese número")
            numCoseno = float(input())
            resCoseno = math.cos(numCoseno)
            print("El coseno es igual a:", resCoseno)

        elif opCoseno == 2:
            # Arco Coseno
            print("Ingrese número")
            numArcCoseno = float(input())
            resArcCoseno = math.acos(numArcCoseno)
            print("El arco coseno es igual a:", resArcCoseno)

        else:
            print("Opción no válida")

    elif opcion == 9:
        # Operación de tangente y arco
        print("Seleccione qué operación desea")
        print("(1) Tangente")
        print("(2) Arco Tangente")
        opTangente = int(input())

        if opTangente == 1:
            # Tangente
            print("Ingrese número")
            numTangente = float(input())
            resTangente = math.tan(numTangente)
            print("La tangente es igual a:", resTangente)

        elif opTangente == 2:
            # Arco Tangente
            print("Ingrese número")
            numArcTangente = float(input())
            resArcTangente = math.atan(numArcTangente)
            print("El arco tangente es igual a:", resArcTangente)

        else:
            print("Opción no válida")

    elif opcion == 10:
        # Operación de ecuación de segundo grado
        print("Ingrese el valor de a")
        a = float(input())
        print("Ingrese el valor de b")
        b = float(input())
        print("Ingrese el valor de c")
        c = float(input())
        d = b * b - (4 * a * c)
        di = 2 * a

        if d <= 0 or di == 0:
            print("No es posible la resolución")
        else:
            x1 = (-b + math.sqrt(d)) / di
            x2 = (-b - math.sqrt(d)) / di
            print("x1 es igual a:", x1)
            print("x2 es igual a:", x2)

    else:
        print("Opción no válida")

    print("Desea continuar? (Si, escriba S; No, escriba N)")
    continuar = input()

    if continuar.lower() == "n":
        repe = 2
```
